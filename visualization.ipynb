```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
x=iris.data
y=iris.target
```

```python
plt.hist(x[:,1],bins=40)
plt.xlabel("sepeal length")
plt.ylabel("sepeal width")
plt.show()
```

```python
#scatter plot
plt.scatter(x[:,0],x[:,1],c=iris.target)
plt.xlabel("sepeal length")
plt.ylabel("sepeal width")
plt.show()
```

```python
#line plot
plt.plot(x[:,1])
plt.xlabel("index")
plt.ylabel("sepeal length")
plt.show()
```

```python
#average barplot
avg=[x[y==i,0].mean() for i in range(3)]
avg
```

```python
avg=[]
for i in range(3):
    avg.append(x[y==i,0].mean())

        
```

```python
tar_name=iris.target_names
tar_name
```

```python
plt.bar(tar_name,avg)
plt.xlabel("name")
plt.ylabel("mean")
plt.show()
```
