```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
```

```python
X = np.array([[-2, 3, -4, 5, -6],
              [1, -2, 3, -4, 5],
              [-5, 4, -3, 2, -1],
              [6, -5, 4, -3, 2],
              [-1, 2, -3, 4, -5]], dtype=np.int32)
y = np.array([[1], [-1], [1], [-1], [1]], dtype=np.int32)
```

```python
model = keras.Sequential([
    # Input layer
    layers.Input(shape=(5,)),
    # Hidden layer 1
    layers.Dense(4, activation="relu"),
    # Hidden layer 2
    layers.Dense(3, activation="relu"),
    # Hidden layer 3
    layers.Dense(2, activation="relu"),
    # Output layer
    layers.Dense(1, activation="relu")
])
```

```python
model.compile(optimizer="adam", loss="mse")
```

```python
# X = np.array([2,5,3,4,6],dtype=np.float32
# x=X.reshape(-1,1)
# y = np.array([1,-1,-1,-1,-1],dtype=np.float32)
# # Y=y.reshape(-1,1)
# X = np.array([[-2, 3, -4, 5, -6],
#               [1, -2, 3, -4, 5],
#               [-5, 4, -3, 2, -1],
#               [6, -5, 4, -3, 2],
#               [-1, 2, -3, 4, -5]], dtype=np.int32)
# y = np.array([[1], [-1], [1], [-1], [1]], dtype=np.int32)
model.fit(X, y, epochs=100, batch_size=32)
```

```python
import plotly.graph_objs as go
import plotly.offline as pyo

# Define your data
diseased_code = [10, 15, 25, 30, 40]
not_diseased_code = [20, 25, 15, 10, 30]

# Create a trace for diseased code
trace1 = go.Bar(
    x=['Code 1', 'Code 2', 'Code 3', 'Code 4', 'Code 5'],
    y=diseased_code,
    name='Diseased '
)

# Create a trace for not diseased code
trace2 = go.Bar(
    x=['Code 1', 'Code 2', 'Code 3', 'Code 4', 'Code 5'],
    y=not_diseased_code,
    name='Not Diseased '
)

# Create a layout
layout = go.Layout(
    title='Diseased and Not Diseased ',
    xaxis=dict(title='Code'),
    yaxis=dict(title='Number of Occurrences')
)

# Combine the traces into a data list
data = [trace1, trace2]

# Create the figure
fig = go.Figure(data=data, layout=layout)

# Show the plot
pyo.plot(fig, filename='diseased-and-not-diseased-code.html')
```
